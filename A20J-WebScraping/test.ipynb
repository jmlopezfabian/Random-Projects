```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
import os
import glob
import time
import numpy as np
```

```python
url = "https://earthshakira.github.io/a2oj-clientside/server/"
response = requests.get(url + "Ladders.html")
if response.status_code == 200:
    soup = BeautifulSoup(response.text, 'html.parser')
else:
    print("Error de conexion D:")
```

```python
response.status_code
```

```python
tabla_rating = soup.find_all('table')[0]
elementos = tabla_rating.find_all('tr')
elementos
```

```python
tabla = soup.find('table')
elementos = tabla.find_all('tr')

for elemento in elementos:
    info = {
        "Nombre" : [],
        "Enlace" : [],
    }

    tds = elemento.find_all('td')
    if len(tds) >= 3:
        link = url + tds[1].find('a').get('href') 
        nombre = tds[1].find('a').text.replace(" ","") + ".csv"
        response_sub = requests.get(link)
        if(response_sub.status_code == 200):
            soup_sub = BeautifulSoup(response_sub.text, 'html.parser')
            tabla_sub = soup_sub.find_all('table')[1]
            elementos_sub = tabla_sub.find_all('tr')
            for elemento_sub in elementos_sub:
                tds_sub = elemento_sub.find_all('td')
                if(len(tds_sub) >= 3):
                    link_sub = tds_sub[1].find('a').get('href')
                    nombre_sub = tds_sub[1].find('a').text.replace(" ","")
                    print(link_sub, nombre_sub)
                    info["Nombre"].append(nombre_sub)
                    info["Enlace"].append(link_sub)
                    data = pd.DataFrame(info)
                    data.to_csv("data/" + nombre)
        else:
            print("Error de conexion D:")
        
```

```python
ruta_carpeta = "/home/gsu/Random-Projects/A20J-WebScraping/data"
archivos_csv = glob.glob(os.path.join(ruta_carpeta,'*.csv'))

iterador = 0
for archivo in archivos_csv:
    
    if iterador == 1:
        break

    data = pd.read_csv(archivo)
    topics = []
    links = list(data["Enlace"])

    for link in links:
        response_cf = requests.get(link)
        while(response_cf.status_code != 200):
            print(response_cf.status_code)
            response_cf = requests.get(link)
        soup_cf = BeautifulSoup(response_cf.text, 'html.parser')
        temas = soup_cf.find_all('div', class_ = 'roundbox sidebox borderTopRound')[1].find_all('div')[2].find_all('span')[0:-1]
        temas_problema = []
        rating = "-"
        for tema in temas:
            print(tema.text)
            if(tema.text[0] == "*"):
                rating = tema.text
            else:
                temas_problema.append(tema.get_text().replace("\n","").replace("\r","").strip())
        
        topics.append(temas_problema)
    data["Temas"] = topics
    iterador += 1
```
