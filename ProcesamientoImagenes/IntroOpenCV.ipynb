# Trabajando con pixeles

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import math


#Imagen de 100x100 de 3 dimensiones en negro
dimensiones = (100, 100)
imagenOscura = np.zeros((dimensiones[0],dimensiones[1],3), np.uint8)

#Reemplazar el pixel de una imagen

radio = 3
centro = 50
rojo = [255,0,0]

def dibujarCuadrado(imagen: np.array, lado: int, centro: list, color: list = [255, 255, 255]) -> None:
    # Asegurarse de que las coordenadas no estén fuera de los límites de la imagen
    for i in range(lado + 1):
        # Borde superior
        if 0 <= (centro[0] - (lado // 2)) < imagen.shape[0] and 0 <= (centro[1] - (lado // 2) + i) < imagen.shape[1]:
            imagen[centro[0] - (lado // 2), centro[1] - (lado // 2) + i] = color
        # Borde inferior
        if 0 <= (centro[0] + (lado // 2)) < imagen.shape[0] and 0 <= (centro[1] - (lado // 2) + i) < imagen.shape[1]:
            imagen[centro[0] + (lado // 2), centro[1] - (lado // 2) + i] = color
        # Borde derecha
        if 0 <= (centro[0] - (lado // 2) + i) < imagen.shape[0] and 0 <= (centro[1] + (lado // 2)) < imagen.shape[1]:
            imagen[centro[0] - (lado // 2) + i, centro[1] + (lado // 2)] = color
        # Borde izquierda
        if 0 <= (centro[0] - (lado // 2) + i) < imagen.shape[0] and 0 <= (centro[1] - (lado // 2)) < imagen.shape[1]:
            imagen[centro[0] - (lado // 2) + i, centro[1] - (lado // 2)] = color

    return imagen
    

def dibujarCirculo(imagen, radio, centro, color) -> None:

    for i in range(0, 360 + 1):
        x = centro[0] + int(round(radio * math.cos(i * (math.pi / 180)), 0))
        y = centro[1] + int(round(radio * math.sin(i * (math.pi / 180)), 0))

        if 0 <= x < imagen.shape[1] and 0 <= y < imagen.shape[0]:
            imagen[y, x] = color  # Asignar el color al píxel correspondiente

    return imagen

circulo = dibujarCirculo(imagenOscura, 30, [50,50], [0,0,255])
circulo = dibujarCirculo(imagenOscura, 35, [50,50], [0,0,255])
circulo = dibujarCirculo(imagenOscura, 40, [50,50], [0,0,255])
circulo = dibujarCirculo(imagenOscura, 45, [50,50], [0,0,255])
cuadrado = dibujarCuadrado(circulo, 60, [50,50], [255,255,255])
plt.imshow(cuadrado)
plt.axis('off')
#cv2.namedWindow("black", cv2.WINDOW_NORMAL)
#cv2.imshow("black", imagenOscura)

#cv2.waitKey()
#v2.destroyAllWindows()
```
